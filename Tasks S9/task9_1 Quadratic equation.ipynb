```python
f(x) = 5x^2 + 10x - 30
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
a, b, c = 5, 10, -30
```

```python
def f(x):
    function = a*x**2 + b*x + c
    return function
```

```python
def root(a, b, c):
    dscrt = b**2 - 4*a*c
    if dscrt > 0:
        x1 = (-b + math.sqrt(dscrt))/(2*a)
        x2 = (-b - math.sqrt(dscrt))/(2*a)
        return round(x1, 2), round(x2, 2)
    if dscrt == 0:
        x = -b /(2*a)
        return x
    else:
        return None
```

```python
print(root(a, b, c))
```

```python
x = np.arange(-10, 10.01, 0.01)
```

```python
y_min = min(f(x))
```

```python
print(f(x))
```

```python
print(y_min)
```

```python
x_min = root(a, b, c - y_min)
```

```python
print(x_min)
```

```python
def show_func():
    plt.plot(x, f(x), 'r', label='Функция f(x)')
    plt.show()
```

```python
show_func()
```

```python
def show_func():
    x_down = np.arange(-10, x_min, 0.01)
    x_up = np.arange(x_min, 10.01, 0.01)
    plt.plot(x_down, f(x_down), 'r', label = 'Убывание f(x)')
    plt.plot(x_up, f(x_up), 'b', label = 'Возрастание f(x)')
    plt.plot(x_min, y_min, 'go', label = 'Точка экстремума ({x_min}, {y_min})')
    plt.grid()
    plt.legend()
    plt.show()
```

```python
show_func()
```

```python
def show_func():
    #x_down = np.arange(-10, x_min, 0.01)
    #x_up = np.arange(x_min, 10, 0.01)
    x_range_down_positive = np.arange(-10.0, root(a, b, c)[1], 0.01)
    x_range_down_negative = np.arange(root(a, b, c)[1], x_min, 0.01)
    x_range_up_negative = np.arange(x_min, root(a, b, c)[0], 0.01)
    x_range_up_positive = np.arange(root(a, b, c)[0], 10.01, 0.01)
    
    plt.title(f'Корни функции: {root(a, b, c)[0]}, {root(a, b, c)[1]}')
    plt.xlabel('Ось Х')
    plt.ylabel('Ось Y')
    plt.grid()
    plt.rcParams['lines.linestyle'] = '-'
    plt.plot(x_range_down_positive, f(x_range_down_positive), 'r', label='Убывание f(x) больше 0')
    plt.rcParams['lines.linestyle'] = '-.'
    plt.plot(x_range_down_negative, f(x_range_down_negative), 'y', label='Убывание f(x) меньше 0')
    plt.plot(x_range_up_negative, f(x_range_up_negative), 'y', label='Возрастание f(x) меньше 0')
    plt.rcParams['lines.linestyle'] = '-'
    plt.plot(x_range_up_positive, f(x_range_up_positive), 'b', label='Возрастание f(x) больше 0')
    plt.plot(x_min, y_min, 'go', label=f'Точка экстремума ({x_min},{y_min})')
    plt.legend()
    plt.show()
```
